```python
# Seleccionar todas las columnas numéricas para análisis
numeric_columns = df.select_dtypes(include=np.number).columns.tolist()
numeric_columns.pop()

n = len(numeric_columns)
cols = 4
rows = n // cols + (n % cols > 0)
plt.figure(figsize=(cols * 4, rows * 3))
plt.style.use('ggplot')

# Definir los límites para las variables específicas
limits = {
    'price': (0, 15000),
    'minimum_nights': (0, 50),
    'number_of_reviews': (0, 100),
    'reviews_per_month': (0, 10),
    'calculated_host_listings_count': (0, 20)
}

for i, variable in enumerate(numeric_columns):
    plt.subplot(rows, cols, i + 1)
    sns.histplot(df[variable], kde=False, color='blue')
    
    # Aplicar límites específicos si existen para la variable
    if variable in limits:
        plt.xlim(limits[variable])
    
    plt.title(variable, fontsize=12)
    plt.xlabel('')

plt.tight_layout()
plt.show()
```

```python
# Variables categóricas
cat_columns = ['neighbourhood', 'room_type']

plt.figure(figsize=(30, 21))

for i, variable in enumerate(cat_columns):
    plt.subplot(4, 4, i + 1)
    
    if variable == 'neighbourhood':
        top_15_neighbourhoods = df[variable].value_counts(ascending=False).head(15).index
        df_top_15 = df[df[variable].isin(top_15_neighbourhoods)]
        order = top_15_neighbourhoods
    else:
        df_top_15 = df
        order = df[variable].value_counts(ascending=False).index
    
    ax = sns.countplot(x=df_top_15[variable], data=df_top_15, order=order, palette='viridis')
    
    # Agregar anotaciones de porcentaje en las barras
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height() / len(df_top_15[variable]))
        x = p.get_x() + p.get_width() / 2
        y = p.get_y() + p.get_height()
        ax.annotate(percentage, (x, y), ha='center', va='bottom', fontsize=12)
    
    # Rotar las etiquetas del eje x para mejorar la legibilidad
    plt.xticks(rotation=90)
    plt.title(f'top 15 {variable}' if variable == 'neighbourhood' else variable, fontsize=14)

# Ajustar el diseño para evitar superposiciones
plt.tight_layout()
plt.show()
```
